## Volume Setup in Unity Catalog

The commands below configure the working environment in **Databricks Unity Catalog**,
defining where files will be stored.

- `USE CATALOG workspace` — selects the catalog named *workspace*, which is the default
  catalog available in Databricks Community/free-tier environments.

- `USE SCHEMA default` — selects the schema (database) *default* inside the catalog,
  which is the namespace where objects will be created.

- `CREATE VOLUME IF NOT EXISTS spark_dev` — creates a **Volume**, a Unity Catalog object
  used to store non-tabular files (such as CSVs, JSONs, images, etc.) in a managed way.
  The `IF NOT EXISTS` clause ensures the command does not raise an error if the Volume
  already exists.

> After this setup, the Volume will be accessible at the path
> `/Volumes/workspace/default/spark_dev/`

```python
spark.sql("USE CATALOG workspace")
spark.sql("USE SCHEMA default")
spark.sql("CREATE VOLUME IF NOT EXISTS spark_dev")
```
